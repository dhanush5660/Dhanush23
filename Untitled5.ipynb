```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# 1. Download Stock Price Data
def download_stock_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    return data['Close'].values.reshape(-1, 1)

# 2. Preprocess the Data
def preprocess_data(data):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    return scaled_data, scaler

# 3. Prepare Time Series Data for LSTM
def create_dataset(dataset, time_step=10):
    X, y = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        X.append(a)
        y.append(dataset[i + time_step, 0])
    return np.array(X), np.array(y)

# 4. Build the LSTM Model
def build_lstm_model(time_step, feature_size=1):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(time_step, feature_size)))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# 5. Train and Evaluate the Model
def train_evaluate_model(model, X_train, y_train, X_test, y_test, epochs=50, batch_size=32):
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)
    train_loss = model.evaluate(X_train, y_train, verbose=0)
    test_loss = model.evaluate(X_test, y_test, verbose=0)
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
    return model

# 6. Make Predictions
def make_predictions(model, X_test, scaler):
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    return predictions

# 7. Visualize Results
def plot_results(actual, predicted):
    plt.figure(figsize=(12, 6))
    plt.plot(actual, label='Actual Price')
    plt.plot(predicted, label='Predicted Price')
    plt.title('Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    ticker = 'AAPL'  # Example: Apple Inc.
    start_date = '2020-01-01'
    end_date = '2023-12-31'
    time_step = 60  # Number of previous days to consider for prediction
    split_ratio = 0.8 # Ratio for train-test split

    # Download data
    raw_data = download_stock_data(ticker, start_date, end_date)

    # Preprocess data
    scaled_data, scaler = preprocess_data(raw_data)

    # Split data into training and testing sets
    train_size = int(len(scaled_data) * split_ratio)
    test_size = len(scaled_data) - train_size
    train_data = scaled_data[0:train_size, :]
    test_data = scaled_data[train_size:len(scaled_data), :]

    # Create datasets
    X_train, y_train = create_dataset(train_data, time_step)
    X_test, y_test = create_dataset(test_data, time_step)

    # Reshape input to be [samples, time steps, features] which is required for LSTM
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

    # Build and train the LSTM model
    model = build_lstm_model(time_step)
    model = train_evaluate_model(model, X_train, y_train, X_test, y_test)

    # Make predictions on the test data
    predictions = make_predictions(model, X_test, scaler)

    # Inverse transform the actual test values to original scale
    actual_prices = scaler.inverse_transform(y_test.reshape(-1, 1))

    # Visualize the results
    plot_results(actual_prices, predictions)

    # Predicting the next day's price (example)
    last_sequence = scaled_data[-time_step:]
    last_sequence = np.reshape(last_sequence, (1, time_step, 1))
    predicted_next_scaled = model.predict(last_sequence)
    predicted_next_price = scaler.inverse_transform(predicted_next_scaled)[0][0]
    print(f"\nPredicted next day's price: {predicted_next_price:.2f}")
```
